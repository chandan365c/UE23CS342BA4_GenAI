```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "text-classification",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
titles = [
    "Watch These Impossible tricks!",
    "How to Train a Neural Network Using PyTorch",
    "This One Trick Will Change Your Life Forever",
    "Lecture 5: Linear Regression and Gradient Descent"
]

results = classifier(titles)

label_mapping = {
    'POSITIVE': 'Clickbait',
    'NEGATIVE': 'Informative'
}

for title, result in zip(titles, results):
    predicted_label = label_mapping.get(result['label'], result['label']) # Fallback
    print(f"Title: {title}")
    print(f"Prediction: {predicted_label}, Confidence: {result['score']:.2f}")
    print("-" * 50)
```

```python
def classify_titles(titles_list):
    """
    Classifies a list of titles as 'Clickbait' or 'Informative' using the pre-loaded classifier.

    Args:
        titles_list (list): A list of strings, where each string is a title to classify.

    Returns:
        list: A list of dictionaries, each containing 'title', 'prediction', and 'confidence'.
    """
    global classifier # Access the globally defined classifier

    results = classifier(titles_list)

    label_mapping = {
        'POSITIVE': 'Clickbait',
        'NEGATIVE': 'Informative'
    }

    formatted_results = []
    for title, result in zip(titles_list, results):
        predicted_label = label_mapping.get(result['label'], result['label'])
        formatted_results.append({
            'title': title,
            'prediction': predicted_label,
            'confidence': result['score']
        })
    return formatted_results

# Example usage:
new_titles = [
    "Watch These Impossible tricks!",
    "How to Train a Neural Network Using PyTorch",
    "This One Trick Will Change Your Life Forever",
    "Lecture 5: Linear Regression and Gradient Descent"
]

predictions = classify_titles(new_titles)

print("Classification Results:")
for item in predictions:
    print(f"Title: {item['title']}")
    print(f"Prediction: {item['prediction']}, Confidence: {item['confidence']:.2f}")
    print("-" * 50)
```

```python
print("Enter titles (one per line). Press Enter twice when finished:")
user_titles = []
while True:
    line = input()
    if not line:
        break
    user_titles.append(line)

if user_titles:
    user_predictions = classify_titles(user_titles)
    print("\n--- Your Classification Results ---")
    for item in user_predictions:
        print(f"Title: {item['title']}")
        print(f"Prediction: {item['prediction']}, Confidence: {item['confidence']:.2f}")
        print("-" * 50)
else:
    print("No titles entered.")
```
